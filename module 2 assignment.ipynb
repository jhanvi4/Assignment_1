```python
'''ans.  MUTABLE->Mutable objects are those that can be modified after they are created. This means you can change their content without creating a new object.

Examples:
	•	Lists
	•	Dictionaries
Example:
Creating a mutable list'''
my_list = [1, 2, 3]

# Modifying the list
my_list.append(4)   # Adding an element
print(my_list)      # Output: [1, 2, 3, 4]

my_list[0] = 10     # Changing an element
print(my_list)      # Output: [10, 2, 3, 4]



#Immutable-> Immutable objects cannot be changed once they are created. Any operation that modifies the object will result in the creation of a new object.

#Examples:

#	•	Strings
#	•	Tuples
    # Creating an immutable string
original_string = "Python"

# Trying to modify the string (this will not work)
# original_string[0] = "J"  # This would raise a TypeError

# Instead, create a new string with the desired change
new_string = "J" + original_string[1:]

# Output the original and new strings
print("Original String:", original_string)  # Output: Python
print("New String:", new_string)              # Output: Jython
    
```

```python
#ans.
### 1. Arithmetic Operators

'''These operators are used to perform mathematical operations.

- *Addition (+)*
- *Subtraction (-)*
- **Multiplication (*)**
- *Division (/)*
- *Floor Division (//)*
- *Modulus (%)*
- **Exponentiation (**)**

Example:'''
a = 10
b = 3

print("Addition:", a + b)         # Output: 13
print("Subtraction:", a - b)      # Output: 7
print("Multiplication:", a * b)   # Output: 30
print("Division:", a / b)         # Output: 3.3333
print("Floor Division:", a // b)   # Output: 3
print("Modulus:", a % b)           # Output: 1
print("Exponentiation:", a ** b)   # Output: 1000


### 2. Comparison Operators

'''These operators compare two values and return a Boolean result (True or False).

- *Equal to (==)*
- *Not equal to (!=)*
- *Greater than (>)*
- *Less than (<)*
- *Greater than or equal to (>=)*
- *Less than or equal to (<=)*

Example:'''

x = 5
y = 10

print("Equal to:", x == y)            # Output: False
print("Not equal to:", x != y)        # Output: True
print("Greater than:", x > y)         # Output: False
print("Less than:", x < y)            # Output: True
print("Greater than or equal to:", x >= y)  # Output: False
print("Less than or equal to:", x <= y)      # Output: True


### 3. Logical Operators

'''These operators are used to combine conditional statements.

- *Logical AND (and)*
- *Logical OR (or)*
- *Logical NOT (not)*

Example:'''
a = True
b = False

print("Logical AND:", a and b)  # Output: False
print("Logical OR:", a or b)    # Output: True
print("Logical NOT:", not a)     # Output: False


### 4. Assignment Operators

'''These operators are used to assign values to variables.

- *Assign (=)*
- *Add and assign (+=)*
- *Subtract and assign (-=)*
- **Multiply and assign (*=)**
- *Divide and assign (/=)*
- *Floor divide and assign (//=)*
- *Modulus and assign (%=)*
- **Exponentiate and assign (**=)**

Example:'''

x = 5
x += 3    # Equivalent to x = x + 3
print("Add and assign:", x)  # Output: 8

x -= 2    # Equivalent to x = x - 2
print("Subtract and assign:", x)  # Output: 6

x *= 2    # Equivalent to x = x * 2
print("Multiply and assign:", x)  # Output: 12

x /= 4    # Equivalent to x = x / 4
print("Divide and assign:", x)  # Output: 3.0


### 5. Bitwise Operators

'''These operators perform operations on bits and are useful for low-level programming.

- *Bitwise AND (&)*
- *Bitwise OR (|)*
- *Bitwise NOT (~)*
- *Bitwise XOR (^)*
- *Left Shift (<<)*
- *Right Shift (>>)*

Example:'''
x = 5  # Binary: 0101
y = 3  # Binary: 0011

print("Bitwise AND:", x & y)       # Output: 1 (Binary: 0001)
print("Bitwise OR:", x | y)        # Output: 7 (Binary: 0111)
print("Bitwise NOT:", ~x)           # Output: -6 (Inverts bits)
print("Bitwise XOR:", x ^ y)       # Output: 6 (Binary: 0110)
print("Left Shift:", x << 1)       # Output: 10 (Binary: 1010)
print("Right Shift:", x >> 1)      # Output: 2 (Binary: 0010)


### 6. Identity Operators

'''These operators are used to check if two variables refer to the same object in memory.

- Is (is)
- is not (is not)

Example:'''

a = [1, 2, 3]
b = a
c = list(a)

print("a is b:", a is b)         # Output: True (both refer to the same object)
print("a is c:", a is c)         # Output: False (c is a different object)
print("a is not c:", a is not c) # Output: True


### 7. Membership Operators

'''These operators test for membership in sequences (like lists, tuples, or strings).

- *In (in)*
- *Not in (not in)*

Example:'''

my_list = [1, 2, 3, 4, 5]

print("3 in list:", 3 in my_list)        # Output: True
print("6 in list:", 6 in my_list)        # Output: False
print("5 not in list:", 5 not in my_list)  # Output:
```

```python
'''ans. Typecasting in Python refers to the conversion of one data type to another. This can be done explicitly or implicitly. Python provides several built-in functions for typecasting, allowing you to convert variables from one type to another, such as integers, floats, strings, and more.


Converting String to Integer and Float:'''

num_str = "25"
num_int = int(num_str)    # Convert string to integer
num_float = float(num_str) # Convert string to float

print(num_int)    # Output: 25
print(num_float)  # Output: 25.0
print(type(num_float))
print(type(num_int))


#	2.	Converting Integer to String:

age = 30
age_str = str(age)  # Convert integer to string

print(age_str)      # Output: "30"
print(type(age_str)) # Output: <class 'str'>


#	3.	Converting List to String:

my_list = [1, 2, 3]
my_list_str = str(my_list)  # Convert list to string

print(my_list_str)     # Output: "[1, 2, 3]"
print(type(my_list_str)) # Output: <class 'str'>
```

```python
'''ans. Conditional statements in Python allow you to execute specific blocks of code based on certain conditions. The most common conditional statements are if, elif, and else. These statements help control the flow of execution in a program by allowing you to make decisions based on the evaluation of boolean expressions.


if condition:
    # block of code executed if condition is true
elif another_condition:
    # block of code executed if another_condition is true
else:
    # block of code executed if none of the conditions are true


Examples

 1. Using if Statement

Example:'''

age = 18

if age >= 18:
    print("You are eligible to vote.")



#### 2. Using if and else Statements

#Example:


age = 16

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")



#### 3. Using elif Statement

#Example:


marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
else:
    print("Grade: F")



#### 4. Nested Conditional Statements

#Example:

number = 15

if number > 0:
    print("The number is positive.")
    if number % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is non-positive.")




#### 5. Using Logical Operators

#Example:
    
temperature = 30
humidity = 70

if temperature > 25 and humidity > 60:
    print("It's a hot and humid day.")
else:
    print("The weather is comfortable.")


```

```python
'''ans. For Loop

The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range). It allows you to execute a block of code for each item in the sequence.

 Use Cases:
- Iterating over elements in a collection.
- Performing a fixed number of iterations.
- Looping through a range of numbers.

Example: Iterating over a List'''


fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)



#Example: Using range()

for i in range(5):  # Iterates from 0 to 4
    print(i)



### 2. While Loop

'''The while loop repeatedly executes a block of code as long as a specified condition is True. It is useful when the number of iterations is not known beforehand.

 Use Cases:
- Running a loop until a specific condition is met.
- Implementing countdowns or timers.
- Reading input until valid data is provided.

Example: Basic While Loop'''
    
count = 0

while count < 5:
    print("Count is:", count)
    count += 1  # Incrementing count



#Example: User Input Validation
    
password = ""

while password != "secret":
    password = input("Enter the password: ")

print("Access granted.")


### 3. Nested Loops

'''You can use loops inside other loops. This is useful for iterating through multi-dimensional data structures (like lists of lists).

 Use Cases:
- Working with matrices or 2D arrays.
- Generating combinations of items.

Example: Nested For Loop'''

for i in range(3):  # Outer loop
    for j in range(2):  # Inner loop
        print(f"i={i}, j={j}")



### 4. Loop Control Statements

'''Python provides control statements that can alter the flow of loops:

- *break*: Terminates the loop.
- *continue*: Skips the current iteration and continues to the next.
- *pass*: A placeholder that does nothing.

Example: Using break'''

for i in range(10):
    if i == 5:
        break  # Exit the loop when i is 5
    print(i)



#Example: Using continue

for i in range(5):
    if i == 2:
        continue  # Skip the iteration when i is 2
    print(i)



#Example: Using pass

for i in range(3):
    if i == 1:
        pass  # Placeholder, does nothing
    print(i)


```
